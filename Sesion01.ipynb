# Sesión 01: Estimados de Locación y Variabilidad

## Datos estructurados

Las variables cuantitativas y cualitativas son términos que se utilizan comúnmente en estadística y se relacionan estrechamente con los conceptos de datos numéricos y categóricos, respectivamente, aunque con algunas sutilezas en su definición:

1. **Variables Cuantitativas:** Corresponden a los datos numéricos. Son variables que se pueden medir en términos de números. Se dividen en dos subcategorías:
   - **Discretas:** Valores que toman números enteros o recuentos, como el número de hijos, automóviles, etc.
   - **Continuas:** Valores que pueden tomar cualquier número dentro de un rango, incluidos decimales, como peso, altura, temperatura.

2. **Variables Cualitativas:** Corresponden a los datos categóricos. Son variables que describen cualidades o características y no pueden expresarse legítimamente en términos numéricos. También se dividen en dos subcategorías:
   - **Nominales:** Datos que no tienen un orden inherente, como género, color, nacionalidad.
   - **Ordinales:** Datos que tienen un orden o jerarquía, pero donde las diferencias no son matemáticamente medibles, como calificaciones (bueno, regular, malo), niveles educativos.

En pocas palabras:
- **Datos Numéricos = Variables Cuantitativas:** Ambos términos se refieren a datos que se expresan en números y se pueden cuantificar.
- **Datos Categóricos = Variables Cualitativas:** Ambos términos se refieren a datos que describen categorías o cualidades y no se expresan naturalmente en números.

Sin embargo, en la práctica, estos términos suelen utilizarse de manera intercambiable. La diferencia más importante es entender si los datos que estás analizando representan cantidades numéricas o categorías cualitativas, ya que esto determinará el tipo de análisis estadístico que puedes realizar.

Algunos ejemplos para ilustrar mejor la diferencia entre datos numéricos (variables cuantitativas) y datos categóricos (variables cualitativas).

### Datos Numéricos (Variables Cuantitativas)

1. **Discretos:**
   - **Número de habitaciones en una casa:** 2, 3, 4, etc.
   - **Cantidad de estudiantes en una clase:** 20, 25, 30, etc.
   - **Número de libros leídos en un año:** 1, 5, 10, etc.

2. **Continuos:**
   - **Altura de las personas:** 1.65 m, 1.75 m, 1.82 m, etc.
   - **Peso:** 55.5 kg, 70.2 kg, 80 kg, etc.
   - **Distancia recorrida en un viaje:** 120.5 km, 256.75 km, 312 km, etc.

### Datos Categóricos (Variables Cualitativas)

1. **Nominales:**
   - **Tipo de vehículo:** automóvil, motocicleta, camión, bicicleta.
   - **Color de una camisa:** rojo, azul, verde, amarillo.
   - **Marcas de teléfonos móviles:** Samsung, Apple, Huawei, Xiaomi.

2. **Ordinales:**
   - **Nivel de satisfacción del cliente:** insatisfecho, neutral, satisfecho, muy satisfecho.
   - **Grados académicos:** primaria, secundaria, preparatoria, universidad.
   - **Tallas de ropa:** pequeña, mediana, grande, extra grande.

En pocas palabras:
- Los **datos numéricos** se refieren a medidas o cantidades que pueden ser contadas o medidas y a menudo se utilizan en operaciones matemáticas.
- Los **datos categóricos** se utilizan para clasificar información en grupos o categorías, donde el énfasis está en la cualidad más que en la cantidad.

## Ejemplo 1: Estimados de Locación

Las medidas de tendencia central son estadísticas que resumen un conjunto de datos mediante un valor representativo. Las tres medidas más comunes son:

1. **Media**: También conocida como el promedio. Se calcula sumando todos los valores del conjunto de datos y luego dividiendo esa suma por el número total de valores. La media es útil para datos que no tienen valores extremadamente altos o bajos que podrían distorsionar el promedio.

2. **Mediana**: Es el valor que se encuentra justo en el medio del conjunto de datos cuando estos están ordenados de menor a mayor. Si hay un número par de valores, la mediana es el promedio de los dos valores centrales. La mediana es especialmente útil para datos con valores extremos, ya que no se ve tan afectada por ellos como la media.

3. **Moda**: Es el valor que aparece con mayor frecuencia en el conjunto de datos. Puede haber más de una moda en un conjunto de datos o incluso ninguna, en caso de que todos los valores aparezcan con la misma frecuencia. La moda es útil para datos categóricos donde no se pueden calcular medias o medianas.

## Ejemplo 2: Media Truncada

**Media Truncada: Un Estimado de Ubicación Robusto**

---

La media truncada es más robusta que el promedio o la mediana, lo que la hace menos sensible a los valores atípicos. Para calcular la media truncada, sigue estos pasos:

1. **Ordenamiento**: Ordena el conjunto de datos de forma ascendente.

2. **Decisión de Truncamiento**: Elige qué porcentaje de los datos truncar. Usualmente, esto varía entre el 5% y el 25%.

3. **Truncamiento**: Divide el porcentaje elegido en dos y elimina esa proporción de los datos tanto del inicio como del final de la secuencia. Por ejemplo, con un truncamiento del 5%, elimina el 2.5% más bajo y el 2.5% más alto de los datos.

4. **Cálculo**: Calcula el promedio con los datos que quedan.

---

## Ejemplo 3: Desviación estándar

**Para obtener la desviación estándar, se llevan a cabo los siguientes pasos:**

---

1. **Obtener el promedio de nuestros datos:** Calcula la media aritmética de los valores del conjunto.

2. **Calcular las diferencias:** Saca todas las diferencias entre cada valor del conjunto y el valor promedio.

3. **Elevar al cuadrado las diferencias:** Eleva al cuadrado todos los resultados obtenidos en el paso anterior.

4. **Sumar los valores cuadrados:** Suma todos los valores cuadrados obtenidos.

5. **Dividir por el número de valores menos uno:** Divide la suma obtenida entre la cantidad de valores en tu conjunto menos uno (esto se hace en el caso de una muestra de una población, para la población completa no se resta uno).

6. **Calcular la raíz cuadrada del resultado:** Extrae la raíz cuadrada del valor resultante para obtener la desviación estándar.

---

Estos pasos describen el cálculo de la desviación estándar en una muestra. Este proceso es fundamental en estadística para medir la variabilidad o dispersión de un conjunto de datos.

## Ejemplo 4: Estadísticos de Orden

```python
df = pd.read_csv('/content/drive/MyDrive/BEDU/Tecnolochicas/Python_Tecnolochicas_ADP/Datasets/melbourne_housing-clean.csv', index_col=0)
df.head()
```

```python
price_range = df['price'].max() - df['price'].min()
formatted_price_range = f'${price_range:,.2f}'
print(formatted_price_range)
```

```python
formatted_value = f'${df["price"].quantile(0.8):,.2f}'
print(formatted_value)
```

```python
formatted_value = f'${df["price"].quantile(0.25):,.2f}'
print(formatted_value)
```

```python
print(f'Valor mínimo: ${df["price"].min():,.2f}')
print(f'Percentil 10: ${df["price"].quantile(0.1):,.2f}')
print(f'Percentil 25: ${df["price"].quantile(0.25):,.2f}')
print(f'Percentil 50: ${df["price"].median():,.2f}')
print(f'Percentil 75: ${df["price"].quantile(0.75):,.2f}')
print(f'Percentil 90: ${df["price"].quantile(0.9):,.2f}')
print(f'Valor máximo: ${df["price"].max():,.2f}')
```

```python
interquartile_range = df["price"].quantile(0.75) - df["price"].quantile(0.25)
formatted_interquartile_range = f'${interquartile_range:,.2f}'
print(formatted_interquartile_range)
```
